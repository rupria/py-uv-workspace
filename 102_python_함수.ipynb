```python
# 함수의 기본구조
# def 함수이름(parameter_1, parameter_2) :
#         수행할 문장
#         return 결과값
# 
# 
# 함수 호출
#    함수이름(argument_1, argument_2)


#keyword only parameter :오직 키워드를 명시해서만 인자를 전달해야 하는 매개변수이다. 함수 정의 시 별표(*) 기호 뒤에 위치한다.
#
#
# 올바른 호출 방식
# proc(a=10, b=20) → 출력: a: 10, b: 20 
# 잘못된 호출 방식 (에러 발생) 
#proc(10, 20) → TypeError: takes 0 positional arguments...



```

```python
age =int(input('나이: '))
gender = input('성별[남,여]:')

if gender =='여':
    print('여성')
elif gender=='남':
    print('남성')
else:
    print('다시 입력해 주세요. [남, 여]') #다시 입력하는 거면 포크레인인것도 출력 안해줘야지. 19세 포크레인이 뭐니
    

if age >=20:
    print('성인')
else :
    print('청소년')

print(age," ",gender,"입니다.")
        
#else:
#    print('다시 입력해 주세요.[남, 여]')
```
