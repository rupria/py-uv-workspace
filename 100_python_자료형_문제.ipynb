문제6)  
100  이하의 자연수 중에서 2의 배수이거나 3의 배수인 수의 개수를 구하세요.

[실행결과]  
67

```python
# 정답.

len(set(range(2, 101, 2)) | set(range(3, 101, 3)))
# 100 이하의 자연수 중에서
# 2의 배수이거나 3의 배수인 수의 개수를 구하는 문제

# 1. 2의 배수 만들기
# range(2, 101, 2)
# → 2부터 시작해서 101 전까지 2씩 증가
# → 2, 4, 6, 8, ..., 100
two_multiple = set(range(2, 101, 2))

# 2. 3의 배수 만들기
# range(3, 101, 3)
# → 3부터 시작해서 101 전까지 3씩 증가
# → 3, 6, 9, 12, ..., 99
three_multiple = set(range(3, 101, 3))

# 3. 2의 배수이거나 3의 배수인 수 구하기
# | 는 set에서 합집합을 의미한다.
# 즉, 2의 배수 전체와 3의 배수 전체를 합치되,
# 6, 12, 18처럼 중복되는 값은 한 번만 포함한다.
result_set = two_multiple | three_multiple

# 4. 개수 구하기
# len()은 자료 안에 들어 있는 값의 개수를 구한다.
answer = len(result_set)

print(answer)

# [튜플을 선택해 2의 배수, 3의 배수를 만든 후 교집합 시도.]
# z = list(range(1,101))
# 
# a = b = tuple(z)
# a = a[1:100:2]
# b = b[2:100:3]
# 
# print(a.intersection(b))
# print(a & b)
```
