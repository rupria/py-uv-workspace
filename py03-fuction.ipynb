````python
# 파라미터와 리턴값이 있느 함수 정의
def max(x, y):
    if x >= y:
        res = x
    else :
        res = y
    return res # 함수가 수행될때 리턴은 오직 한번만 사용
````

4. positional only
- positional only는 함수를 사용할 때 positional방식으로만 parameter를 쓸 수 있도록 강제하는 방법이다.
- / 의 앞에 오는 parameter들은 positional방식으로 입력할 수 있지만,  keyword방식으로 입력할 수 없다.
- positional only 함수를 정의할 수 있는 것은 python 3.8부터 지원한다. 이전 버전에서는 사전 정의된 positional only 함수는 있지만, 사용자가 만들 수는 없었다.


````python
def proc(a,b,c,/):
    return a,b,c
````

````python
print(proc(4,5,6))
````

````python
print(proc(4,5)) # TypeError: proc() missing 1 required positional argument: 'c' 갯수가 맞아야 합니다.
````

````python
print(proc(4,c=6,b=5))
````

````python
def proc(a,b,/,c,d): # / 기준으로 a,b 무조건 position c,d는 position  또는 keyword 상관없음. 
    return a,b,c,d
````

````python
print(proc(4,5,6,7))
````

````python
print(proc(4,5, d=7,c=6))
````

7. 가변 keyword
- 가변 keyword는 parameter 이름에 '**'를 붙인다.
- keyword방식으로 입력되는 argument의 개수에 상관없이 모두 받아온다.

````python
def proc(**a):
    print(type(a))
    return a
````

````python
def proc(*a,*b): # SyntaxError: * argument may appear only once
    return a,b
````

## return
  - return의 정의
     - 함수 내부의 실행을 종료하고, 호출한 곳으로 결과값을 전달하는 키워드
  - return의 핵심 역할 2가지  
    - 값 반환: 함수에서 계산되거나 처리된 데이터(결과물)를 외부로 내보냄  
    - 함수 종료: return문이 실행되는 즉시 뒤에 코드가 더 있더라도 함수가 끝남
  - 아무것도 리턴하지  않으면 None을 출력함.  None은 아무것도 없다는 의미임.
  - 함수에서 return을 2번 이상 사용할 수 없다.
  - 함수의 결과를 2개 이상으로 받고 싶다면 unpacking해서 받는다.


````python
def proc(a):
    print(a)
````

````python
print(proc(3))
````

````python
def proc(a):
    print(a)
    return # 함수 중간에 강제적으로 실행중인 함수를 빠져나올때 사용. 이거 return 말고 break 도 있지않나, 파이썬은 안쓰던가.
````

````python
print(proc(3))
````

````python
#proc() 함수에서 합계와 평균을 구해서 리턴하는 프로그램을 구현
def proc(*a): # 튜플로 받은 값을 합계와 평균을 구해라
    total = 0 # total = total + k, total*k 는 초기값을 1로 줘야함.
    avg = 0 # avg = total/3

    for v in a:
        total = total + v

    avg = total / len(a)

    return total, avg
````

````python
print(proc(1,2,3,4)) 
````

````python
total, avg = proc(1,2,3,4)
print(f'합계:{total}')
print(f'평균:{avg}')
````
