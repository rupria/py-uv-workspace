````python
## callable
# - 이름 뒤에( )를 붙이는 것을 call(호출)이라고 한다.
# - 이름 뒤에( )를 붙일 수 있는 것을 callable이라고 한다.
# - 함수는 사용할 때 이름 뒤에 괄호를 붙인다.
# - python에서 callable인 것은 3가지 유형이 있다.
#   - (1)function
#   - (2)class(init가 정의되었을 경우)
#   - (3)object(call이 정의되었을 경우)
````
