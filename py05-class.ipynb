 - 이미 존재하는 클래스의 모든 특징을 물려받는 새로운 클래스를 상속이라 한다.

  지칭하는 자식(child class)를 의미한다.

### classmethod
- 첫 번째 인자로 클래스(cls)를 받는다.
- 클래스를 이용해 무언가를 해야할때 사용한다.

```python
class B:
  x=3          # 클래스 변수
  @classmethod
  def aa(cls):  # classmethod는 관례상 cls를 사용한다.
    print('python')


  def bb(self): #self 는 객체 자기 자신입다.
    print('instance')
```

### staticmethod
- 첫 번째 인자로 self, cls를 받지 않는다.
- 클래스나 인스턴스와 관계없는 일반적인 유틸리티 함수를 정의할때 사용한다.
- 독립적인 기능이지만 클래스 안에 넣는 것이 구조적으로 더 나을때 사용한다.





| 항목         | classmethod             | staticmethod |
| ---------- | ----------------------- | ------------ |
| 첫 번째 인자    | `cls` (클래스)             | 없음           |
| 클래스 변수 접근  | 가능                      | 불가능          |
| 인스턴스 변수 접근 | 불가능                     | 불가능          |
| 주 용도       | 클래스 상태를 사용하는 메서드        | 유틸리티 함수 정의   |
| 호출 방식      | 클래스.메서드() 또는 인스턴스.메서드() | 동일           |




```python
class B:
  x = 1


  def xx(self, x): # 인스턴스 메소드
    self.x = x


  @classmethod
  def yy(cls, x):  # 클래스 메소드
    cls.x = x


  @staticmethod # 이거 없어지면 self 로 사용할거임
  def zz(a, b):  # 스테이틱 메소드 a 를 일반 매게변수로 사용
    print(a+b)


  @staticmethod  # 스테이틱 메소드
  def kk():
    print(x) # 스테이트로 선언된 메소드는 클래스에 있는 x =1 을 쓰지 않음
```

```python
class Book: #마치 privit 처럼 사용 가능하게 __를 씁니다.

    def __init__(self, title=None, author=None, page=None):
        self.__title = title
        self.__author = author
        self.__page = page

    def getTitle(self):
        return self.__title
    
    def setTitle(self, title) :
        self.__title = title

    def getAuthor(self):
        return self.__author


    def setAuthor(self, author):
        self.__author = author


    def getPage(self):
        return self.__page


    def setPage(self, page):
        self.__page = page


    def toString(self):
        return f"{self.__title} 의 저자는 {self.__author}이고 {self.__page}페이지로 구성됨"
```

```python
class BookRun :
    def process(self):
        bk1 = Book()
        bk1.setTitle('파이썬의 정석')
        bk1.setAuthor('김파이썬')
        bk1.setPage(150)
        print(bk1.toString())
        # bk1.__title
        # print(bk1._Book_title) # 맹글링 기법 직접적으로 갖고 올수 없는것보단 지양하는 방법임.
        # print(dir(bk1))

        bk2 = Book('파이썬 입문', '박북스', 200) #getter 메서드
        print('책이름 : {}'.format(bk2.getTitle()))
        print('책저자 : {}'.format(bk2.getAuthor()))
        print('페이지 : {}'.format(bk2.getPage()))
        print(bk2.toString())



if __name__ == '__main__':
    bn = BookRun()
    bn.process()
    
```

```python
class Dog:
  count = 0


  def __init__(self, name):
    self.name = name
    Dog.count = Dog.count + 1


  @classmethod
  def get_count(cls):
    return cls.count
  
  @staticmethod 
  def is_valid_name(name):
    # name 의 클래스가 str 이고 길이는 0보다 크면 된다.
    return isinstance(name, str) and len(name) > 0
```

## 상속(inheritance)
 - class에서 이름 뒤에 괄호는 상속이다.
 - python의 모든 객체는 object를 상속받는다.
 - python3부터는 (object)는 생략 가능하다. 즉 지금까지 class를 정의할때 object를 내부적으로 상속하고 있었다.
 - 기본 클래스를 물려받아서 확장해서 사용하는 개념

clas 도형:<br>
 pass

  pass


class 사각형(도형):  # 사각형은 도형이다. is a  <br>
  pas


```python
# 부모클래스
class Parent:
  x = 'parent x'  #클래스 변수
  y = 'parent y'  #클래스 변수


  @classmethod
  def output(cls):
    return cls.x


  def test(self):
     return 10

#자식클래스
class Child(Parent):
  x = 'child x' # 클래스 변수


  def display(self):
    print('display')
    print(Parent.output())
    print(Child.output())
    print(Parent.test(self))
    print(Child.test(self))

```

```python
print(Parent.x)
print(Parent.y)
```

```python
print(Child.x)
print(Child.y)
```

```python
c = Child()
c.display()


# parent x 부모 클래스를갖고옴
# child x 자식클래스를 갖고옴
# 10 고정된 값이 나옴, 변동된게 없어서.
# 10

```

```python
# 부모 클래스 정의
class Animal:
    def __init__(self, name):
        self.name = name


    def make_sound(self):
        # 자식 클래스들이 반드시 재정의(오버라이딩)하도록 유도하는 부모의 기본 메서드
        return "알 수 없는 동물의 소리입니다."


# 자식 클래스 1 (Dog)
class Dog(Animal):
    def make_sound(self):
        # 부모의 make_sound 메서드를 자식에 맞게 완전히 새롭게 덮어씀 (오버라이딩)
        return "멍멍!"


# 자식 클래스 2 (Cat)
class Cat(Animal):
    def make_sound(self):
        # 부모의 make_sound 메서드를 자식에 맞게 완전히 새롭게 덮어씀 (오버라이딩)
        return "야옹~"
```

```python
animal = Animal('동물') # Animal을 animal로 바꿔서 부모 Animal 의 make_sound 를 가져옴 > 가져오고 다른 요소도 추가해도 괜찮나?
print(animal.make_sound())

dog = Dog('강아지')
print(dog.make_sound())

cat = Cat('고양이')
print(cat.make_sound())
```

## overloading(오버로딩)
- parameter에 따라서 다른 함수를 사용하는 기법을 overloading이다.  
- python은 overloading을 지원하지 않고, 뒤에 것이 앞에 것을 덮어쓴다.  


- python은 overloading을 지원하지 않으므로 default value를 많이 만들어 두는 기법을 자주 사용한다
- 위의 overloading예시를 default value를 이용해 python에서 구현하면 다음과 같이 만들 수 있다.

```python
def student(first, last='Mark', age=30):
    print(first)
    print(last)
    print(age)




student('Young')
student('Young', 'Lee')
```

```python
class Base:
  def __init__(self):   # 8
    self.x = 'Base x'   # 9
    self.y = 'Base y'   # 10
    print('Base')       # 11


class A(Base):
  a = 1
  def __init__(self):   # 4
    super().__init__()  # 5
    self.x = 'A x'      # 14
    print('A')          # 15


class B(Base):
   b = 2
   def __init__(self):  # 6
    super().__init__()  # 7
    self.x = 'B x'      # 12
    print('B')          # 13


class C(A, B):
  def __init__(self):   # 1
    print('C init')     # 2
    super().__init__()  # 3          # python 3에서 사용 
    print('C')          # 16
```

```python
class User:
  def __init__(self, name, email):
    self.name = name
    self.email = email


class AdminUser(User):
  def __init__(self, name, email, role): # 
    super().__init__(name, email)       # 부모 생성자를 다시 돌려주고
    self.role = role                    # 이거만씀
```

```python
admin = AdminUser('Alice', 'alice@gmail.com', 'admin')
print(admin.name)
print(admin.email)
print(admin.role)
```

```python
from abc import ABC, abstractmethod


class Shape(ABC): #


    def __init__(self, width, height):
        self.width = width
        self.height = height


    @abstractmethod # Shape를 를 상속받으면 area 를 무조건 오버라이딩해야함. 강제성
    def area(self):
      pass    




class Rect(Shape):
   
    def __init__(self, width, height ):
        super().__init__(width, height)


    def area(self):
      return self.width * self.height  


   
class Triangle(Shape):
   
    def __init__(self, width, height ):
        super().__init__(width, height)


    def area(self):
      return self.width * self.height * 0.5
```

```python
# s = Shape(3,2)
```

```python
r = Rect(2,5)
print(f'사각형 넓이 : {r.area()}')

t = Triangle(2,5)
print(f'사각형 넓이 : {t.area()}')
```
