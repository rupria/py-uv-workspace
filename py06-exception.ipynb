## 예외처리
- 예외처리는 에러가 발생해되 실행을 중단하지 않고 계속 실행하게 해주는 방법이다.
- 예외처리 구조
```
 try:
    실행할 코드
 except:
    오류가 발생했을 때 처리하는 코드
 else:
    오류가 없을 때 처리하는 코드
 finally:
    무조건 처리하는 코드      
```
- 파이썬에서는 선언할 때 오류를 일일이 체크하지 않고, 사용할 때 에러가 발생하면 디버깅하는 스타일이다.
- EAFP
   - 허락보다는 용서를 구하기가 쉽다.(Easier to ask for forgiveness than permission)
   - 출처 : https://docs.python.org/3/glossary.html


- except 뒤에 에러 이름을 넣으면, 해당하는 에러가 발생했을 때만 except를 발생시킨다.
- Built-in Exceptions 예외 계층도 : https://docs.python.org/3/library/exceptions.html#exception-hierarchy  


### try 블록이 정상적으로 실행되지 않으면 else블록은 실행되지 않는다.
### try영역에서 예외이 발생했을 때 : try 실행 -> except 실행 -> finally 실행
### try영역이 정상적으로 수행했을 때 :   try 실행 -> else  실행 -> finally  실행

````python
# try 블록이 정상적으로 실행되지 않으면 else 블록은 실행되지 않는다.
# try 영역에서 예외가 발생했을 때 : try 실행 > except 실행 > finally 실행
# try 영역이 정상적으로 수행했을 때 : try 실행 > else 실행 > finally 실행

try:
    a = int(3)
    b = int('a')
    c = int(4)

except:
    print('숫자가 아닌 문자가 입력되었습니다.')

else :
    print(f'{a}와{b}와{c}의 합:{a+b+c}')

finally:
    print('program end')
````

````python
class Bird:

    def fly(self):
        raise NotImplementedError    # 상속받은 자식클래스에서 fly()을 오버라이딩 해줘야 한다.

class Eagle(Bird):

    def display(self):
        print('display')

    def fly(self):    # fly() 오버라이딩을 안하면 오류 발생
        print('very fast')

eagle = Eagle()
eagle.fly()
````


## 파일 입출력


- 파이썬 표준 함수(내장함수)에는 파일과 관련된 처리를 하는 함수가 기본으로 제공된다.
- 파일은 텍스트 파일과 바이너리 파일로 나뉜다.
- 파일을 열때 open( )함수를 사용한다.
- 파일 입출력 순서
 ① 어플리케이션이 운영체제에게 파일 처리를 API 함수를 통해 의뢰하면, 운영체제가
   요청한 업무를 수행해 주고 그 결과를 어플리케이션에게 돌려줌.
 ② 어플리케이션이 파일 처리 업무를 의뢰하는 과정과 각 과정에서 사용되는 파이썬
    함수






| 단계 | 처리 흐름 | 매칭 소스 코드 | 세부 설명 |
|---|---|---|---|
| ① | 파일 열기 | file = open() | open() 내장함수로 파일을 연다. |
| ② | 파일 읽기/쓰기 | file.read() / file.write() | 얻어진 파일 객체에서 데이터를 읽고 쓴다.(read,write) |
| ③ | 파일 닫기 | file.close() | close() 함수로 객체의 사용을 종료한다. |

파일 읽기/쓰기 함수


| 구분 | 함수(메서드)명 | 반환 및 특징 |
| :---: | :--- | :--- |
| **읽기 (Read)** | `read()` | 파일 전체 내용을 하나의 문자열로 반환 |
| **읽기 (Read)** | `readline()` | 파일의 현재 위치에서 딱 한 줄만 반환 |
| **읽기 (Read)** | `readlines()` | 모든 줄을 읽어 문자열 리스트로 반환 |
| **쓰기 (Write)** | `write()` | 문자열 하나를 파일에 기록 |
| **쓰기 (Write)** | `writelines()` | 문자열 리스트/튜플을 한 번에 기록 |

````python
fread = open('source/sample.txt','r')    # 파일을 읽기로 오픈한다. 읽기 시에는 대상 파일이 필요하다.
# fread = open('source/smple.txt','r' > FileNotFoundError: [Errno 2] No such file or directory: 'source/smple.txt' > 그런 파일이 없어요.
line = fread.read()                      # read 는 파일의 모든 데이터를 읽어온다.                
print(line)
print(type(line))                       #<class 'str'> type 를 뭘 쓰는지 팝업을 봅시다.    
fread.close()                           # 파일 닫기
````

````python
fread = open('source/sample.txt','r')
while True:
    line = fread.readline() #readline()은 파일에서 한 라인씩 읽어옴
    if not line:            # 파일에서 더 이상 읽어올 내용이 없을 시
        break               # 현재 수행 중인 반복문을 빠져나온다.
    
    print(line, end='') # end='' > 빈 줄이 출력되지 않도록 해당 값을 주면 줄 간격이 없어집니다.

fread.close()               #파일 닫기
````
