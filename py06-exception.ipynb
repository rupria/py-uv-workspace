## 예외처리
- 예외처리는 에러가 발생해되 실행을 중단하지 않고 계속 실행하게 해주는 방법이다.
- 예외처리 구조
```
 try:
    실행할 코드
 except:
    오류가 발생했을 때 처리하는 코드
 else:
    오류가 없을 때 처리하는 코드
 finally:
    무조건 처리하는 코드      
```
- 파이썬에서는 선언할 때 오류를 일일이 체크하지 않고, 사용할 때 에러가 발생하면 디버깅하는 스타일이다.
- EAFP
   - 허락보다는 용서를 구하기가 쉽다.(Easier to ask for forgiveness than permission)
   - 출처 : https://docs.python.org/3/glossary.html


- except 뒤에 에러 이름을 넣으면, 해당하는 에러가 발생했을 때만 except를 발생시킨다.
- Built-in Exceptions 예외 계층도 : https://docs.python.org/3/library/exceptions.html#exception-hierarchy  


### try 블록이 정상적으로 실행되지 않으면 else블록은 실행되지 않는다.
### try영역에서 예외이 발생했을 때 : try 실행 -> except 실행 -> finally 실행
### try영역이 정상적으로 수행했을 때 :   try 실행 -> else  실행 -> finally  실행


````python
a =10 / 0 > # 오류가 안떠...? 계속 루프 아이콘이 돌고있네.....
````

````python
# try 블록이 정상적으로 실행되지 않으면 else 블록은 실행되지 않는다.
# try 영역에서 예외가 발생했을 때 : try 실행 > except 실행 > finally 실행
# try 영역이 정상적으로 수행했을 때 : try 실행 > else 실행 > finally 실행

try:
    a = int(3)
    b = int('a')
    c = int(4)

except:
    print('숫자가 아닌 문자가 입력되었습니다.')

else :
    print(f'{a}와{b}와{c}의 합:{a+b+c}')

finally:
    print('program end')
````

### pickle 모듈이란?  
- 파이썬 객체 구조를 그대로 바이트 스트림으로 변환(직렬화)하고, 이를 다시 파이썬 객체로 복원(역직렬화)하는 파이썬 고유의 표준 모듈


- 핵심 함수  
 ①pickle.dump(obj, file): 객체를 바이너리 파일에 저장  
 ②pickle.load(file): 바이너리 파일로 부터 객체를 복원


- ⚠치명적인 보안 경고 (중요)  
①pickle은 신뢰할 수 없는 데이터(인터넷에서 다운로드한 파일 등)를 역직렬화할 때    악성 코드가 자동으로 실행될 수 있는 심각한 보안 취약점이 있다.  
②따라서 외부 통신용 데이터 저장에는 절대로 사용을 지양해야 한다.

###  json (웹 표준 및 안전한 대체재)
- json 모듈?
  ① 텍스트 기반(문자열)으로 저장되어 사람이 직접 읽고 수정할 수 있음<br>
  ②자바스크립트, 자바, C++ 등 다른 프로그래밍 언어와 데이터 교환 시 표준으로 사용<br>
  ③텍스트 형식이므로 pickle과 같은 악성 코드 실행 위험이 없어 매우 안전함<br>
- 핵심 암기 공식: 's' = String (문자열)<br>
  - s가 없는 함수 (dump, load)   →  '파일(File)' 객체를 직접 읽고 쓸 때 사용<br>
  - s가 있는 함수 (dumps, loads) →  '문자열(String)' 데이터를 직접 다룰 때 사용<br>


- 4대 함수 기능 요약
  ① json.dump()  : 파이썬 객체를 'JSON 파일'로 저장 (입력)<br>
  ② json.dumps() : 파이썬 객체를 'JSON 문자열'로 변환 (변환)<br>
  ③ json.load()  : 'JSON 파일'을 파이썬 객체로 읽어옴 (출력)<br>
  ④ json.loads() : 'JSON 문자열'을 파이썬 객체로 변환 (복원)<br>


- json.dump() / dumps() 핵심 옵션<br>
 - 파이썬 객체를 JSON 형식으로 변환할 때 가독성, 정렬, 한글 처리, 데이터 예외 등을
    세밀하게 제어하기 위해 인자(Argument) 값을 추가할 수 있다.
    ① indent : 들여쓰기 공백 수 지정 (가독성 향상)<br>

    ④ separators : 데이터 구분 기호 변경 (용량 압축)<br>
    ⑤ default : 직렬화 불가능한 객체의 처리 함수 지정 (에러 방지)<br>




***문자열 전용 함수 (dumps / loads) 예제***
- 파일 입출력(I/O) 과정 없이, 메모리 상에서 파일 형식을 가진 문자열로만 주고받을 때 사용합니다. (웹 통신, API 전송 등)


````python
import json

data= {'name' : 'kim', 'age': 25}

json_string = json.dumps(data)      # dict -> str
print(json_string)                  # {'name' : 'kim', 'age': 25}
print(type(json_string))            # str
````

````python
json_dict = json.loads(json_string)
print(json_dict)                        # {'name' : 'kim', 'age': 25}
print(type(json_dict))                  # <class 'dict'>
````

````python
import json
from datetime import datetime

data = {
        'c_group': '인공지능팀',
        'a_id': 1004,
        'd_data':[1,2,3],
        'b_date': datetime.now()
    }


print(datetime.now())


#datatime 객체를 문자열로 변환해 주는 함수.
def json_default_handler(obj):
    if isinstance(obj, datetime):
        return obj.strftime('%Y-%m-%d %H:%m:%S')





# 모든 옵션을 집약하여 호출
result = json.dumps(
        data,
        indent=4,
        ensure_ascii=False,
        sort_keys=True,
        default= json_default_handler
    )
print(result)
````

### csv (Comma-Separated Values)
  - 표 형태의 데이터를 쉼표(,)로 구분하여 저장하는 표준 모듈
  - 엑셀(Excel)이나 데이터 분석 라이브러리(Pandas)와 완벽하게 연동됨
  - newline='' 옵션: 윈도우 환경 등에서 데이터 저장 시 줄바꿈이 중복되는 현상을 방지
  - encoding='utf-8-sig': 한글 데이터가 엑셀(Excel)에서 깨지지 않도록 인코딩 적용


````python
import csv

header = ['학번','이름','전공']
students =[['202601','홍길동','컴퓨터공학'],['202602','이순신','데이터사이언스'],['202603','유관순','인공지능학'],
]

#csv로 데이터 저장
# 저장 시 새로운 라인이 추가되지 않도록 newline='' 로 진행해야 새로운 라인을 추가하지 않는다.
with open('source/records.csv','w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(header)           # 헤더작성(제목)
    writer.writerows(students)        # 전체 데이터 작성


#csv 파일 내용 읽어오기
with open('source/records.csv','r') as f:
    reader = csv.reader(f)
    print(reader)
    for row in reader:
    #    print(now)
    print(f'학번:{row[0]} | 이름 :{row[1]} | 전공:{row[2]}')
````
